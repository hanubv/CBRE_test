```python
import numpy as np
import pandas as pd
import glob
import os
from datetime import datetime as dt
from pandas.tseries.offsets import *
```

```python
path = "/content/drive/MyDrive/CBRE"
all_files = glob.glob(path+"/*.csv")

full_data = []
for file in all_files:
    cbre_df = pd.read_csv(file, header = 0, index_col = None)
    full_data.append(cbre_df)

cbre_df2 = pd.concat(full_data, axis= 0, ignore_index= None, sort = False)
```

```python
cbre_df2.head()
```

```python
print(cbre_df2.shape)
```

```python
cbre_df2 = cbre_df2.drop_duplicates()
cbre_df2
```

```python
cbre_df2.isnull()
```

```python
cbre_df2 = cbre_df2.replace(to_replace = np.nan, value = 1337)
cbre_df2
```

```python
cbre_df2["Acctg Date"] = pd.to_datetime(cbre_df2["Acctg Date"])
cbre_df2["Date"] = pd.to_datetime(cbre_df2["Date"])
cbre_df2.info()
```

```python
cbre_df2['no_of_diff'] = (cbre_df2['Acctg Date']-cbre_df2["Date"]).dt.days
cbre_df2
```
